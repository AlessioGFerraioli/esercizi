```python
import pandas as pd
import numpy as np

# Larger employee dataset
data = {
'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ian', 'Jack', 'Kara', 'Liam'],
'Department': ['HR', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'HR', 'Finance', 'IT'],
'Salary': [50000, 60000, 70000, 52000, 58000, 72000, 55000, 62000, 75000, 51000, 63000, 76000],
'Experience': [2, 5, 8, 3, 4, 10, 4, 6, 12, 1, 7, 15]
}
df = pd.DataFrame(data)
```

```python
# find the average salary of employees in each department

df.groupby('Department')['Salary'].mean()
```

```python
# find the employees with higher than the average salary

df[df['Salary'] > df['Salary'].mean()]
```

```python
# add the rank of the employees based on their salary

df['Rank'] = df['Salary'].rank(ascending=False)
```

```python
df
```

```python
# group the employees by department and show them sorted by rank

def sort_by_rank(group_to_sort):
    return group_to_sort.sort_values('Rank')

df.groupby('Department').apply(sort_by_rank)
```
