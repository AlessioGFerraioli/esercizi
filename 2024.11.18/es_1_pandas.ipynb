```python
#1. caricare i dati in un dataframe autogenerandoli casualmente

import pandas as pd
import numpy as np
import random 
import string 

def generate_name():
    N = random.randint(5,8)
    name = (''.join(random.choices(string.ascii_lowercase, k=N)))
    name = name[0].upper() + name[1:]
    return name
# creare un dataframe casuale

data = {"nome" : []}

len_data = 20
for i in range(len_data):
    data["nome"].append(generate_name())
  

df = pd.DataFrame(data) 
```

```python
df.head()
```

```python
df["Città"] = [generate_name() for _ in range(len_data)]
```

```python
df
```

```python
eta = np.random.randint(18, 65, size=len(df))
df["età"] = eta
```

```python
salari = np.random.randint(10000, 100000, size=len(df))

df["salario"] = salari 
```

```python
# 5 identify and remove any duplicate rows

df.drop_duplicates(inplace=True)
```

```python
# 6 address missing values in the dataframe by replacing them with the median of the column 

# i create a nan value in the salario column
df.loc[random.randint(0, len(df)-1), "salario"] = np.nan

df.describe()


```

```python
# fill missing values with median
df["salario"].fillna(df["salario"].median(), inplace=True)
```

```python
# 7 add a new column called "categoria eta" which classifies people as young, adult or senior based on their age
for i in range(len(df)):
    if df.loc[i, "età"] < 30:
        df.loc[i, "Categoria eta"] = "giovane"
    elif df.loc[i,"età"] < 60:
        df.loc[i,"Categoria eta"] = "adulto"
    else:
        df.loc[i, "Categoria eta"] = "senior"

df.head()
```

```python
# export the dataframe to a csv file

df.to_csv("df_prova.csv", index=False)
```
