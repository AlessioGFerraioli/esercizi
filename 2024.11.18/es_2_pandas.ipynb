## Manipolazione e aggregazione dei dati

```python
import pandas as pd
import numpy as np
import random 
import string 

def generate_name():
    N = random.randint(5,8)
    name = (''.join(random.choices(string.ascii_lowercase, k=N)))
    name = name[0].upper() + name[1:]
    return name

nomi_prodotto_possibili = []

for i in range(1, 15):
    nomi_prodotto_possibili.append(generate_name())
  
# creare un dataframe casuale
```

```python
data = {'prodotto': []}

len_data = 50
for i in range(len_data):
    data["prodotto"].append(random.choice(nomi_prodotto_possibili))
  

df = pd.DataFrame(data) 
```

```python
df.head()
```

```python
df["quantita"] = np.random.randint(1, 100, len_data)
df["prezzo unitario"] = np.random.randint(1, 100, len_data)

citta_possibili = ["roma", "milano", "torino", "firenze", "bologna", "napoli"]
citta = []
for i in range(len_data):
    citta.append(random.choice(citta_possibili))
df["citta"] = citta
```

```python
df.head(20)
```

```python
# raggruppare i dati per prodotto e calcolare la somma delle quantità
df_grouped_by_product = df.drop("citta", axis=1).groupby("prodotto").sum()
```

```python
df_grouped_by_product.drop("citta", axis=1, inplace=True)
```

```python
df_grouped_by_product
```

```python
# trova il prodotto piu venduto in termini di quantita

prodotto_piu_venduto = df_grouped_by_product.sort_values("quantita", ascending=False).index[0]
print(f"il prodotto piu venduto è : {prodotto_piu_venduto}")
```

```python
# identificare le citta con il maggior numero di prodotti venduti

df_grouped_by_city = df.drop("prodotto", axis=1).groupby("citta").sum()
df_grouped_by_city.sort_values("quantita", ascending=False).head(5)
```

```python
df["prezzo totale"] = df["quantita"] * df["prezzo unitario"]
df.head(30
        )
```

```python
df_vendite_piu_di_2000 = df[df["prezzo totale"] > 2000]
df_vendite_piu_di_2000.head(20)
```

```python
# ordina il dataframe originale per prezzo totale in modo decrescente

df.sort_values("prezzo totale", ascending=False, inplace=True)
df.head(20)
```

```python
# contare il numero di vendite per ogni citta

df_vendite_per_citta = df.groupby("citta").count()
df_vendite_per_citta.drop("prodotto", axis=1, inplace=True)
#df_vendite_per_citta.sort_values("quantita", ascending=False).head(5)
```

```python
df_vendite_per_citta = df.groupby("citta").sum()

df_vendite_per_citta.drop("prodotto", axis=1, inplace=True)
df_vendite_per_citta.head()
```
