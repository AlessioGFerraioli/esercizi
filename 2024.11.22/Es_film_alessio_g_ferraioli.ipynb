## Weighted User Ranking

Ho bisogno di definire una metrica di ranking che dipenda sia dal voto che dal numero di voti.
La metrica di ranking che utilizzo è la stessa usata da IMDB per le proprie classifiche, ed è definita come segue:

weighted rating (WR) = (v ÷ (v+m)) × R + (m ÷ (v+m)) × C , where:

* R = average for the movie (mean) = (Rating)
* v = number of votes for the movie = (votes)
* m = minimum votes required to be listed in the Top 250 (currently 3000)
* C = the mean vote across the whole report 

Notiamo come questa metrica oltre a calcolare un punteggio unico per un film, basandosi sia sul voto che sul numero di voti, esclude a priori tutti i film che hanno avuto pochi voti (meno di 3000)

```python
def weighted_user_rating(row, C, m=3000):
    '''
    weighted rating formula

    R : average rating for the movie
    v : number of votes for the movie
    C : mean vote across the whole dataset
    m : minimum votes required to be listed in the Top 250
    
    '''
    R = row['vote_average']
    v = row['vote_count']
    
    if v < m:
        return 0
    else:
        return (v/(v+m))*R + (m/(v+m))*C
    

# Creo una nuova colonna "weighted_user_rating" con il punteggio weighted user rating
C = df['vote_average'].mean()
df['weighted_user_rating'] = df.apply(lambda row: weighted_user_rating(row, C), axis=1)
```

## Weighted IMDB rating

Calcolo lo stesso tipo di media pesata anche per il voto IMDB

```python
def weighted_IMDB_rating(row, C, m=3000):
    '''
    weighted rating formula

    R : average rating for the movie
    v : number of votes for the movie
    C : mean vote across the whole dataset
    m : minimum votes required to be listed in the Top 250
    
    '''
    R = row['averageRating']
    v = row['numVotes']
    
    if v < m:
        return 0
    else:
        return (v/(v+m))*R + (m/(v+m))*C


# Creo una nuova colonna "weighted_IMDB_rating" con il punteggio weighted user rating
C = df['averageRating'].mean()
df['weighted_IMDB_rating'] = df.apply(lambda row: weighted_IMDB_rating(row, C), axis=1)
```

Il numero di film è calato a partire dal 2020, probabilmente a causa della pandemia.

Notiamo come il __numero di voti__ dopo il 2020 __cala più velocemente del numero di film__ dopo il 2020: probabilmente questo è indice del fatto che non solo il numero di voti è calato per il minor numero di film, ma anche perché i film prodotti sono così recenti da __non essere stati ancora visti__ (e votati) da molti utenti.

Possiamo anche notare come ci siano dei __picchi relativi in corrispondenza ad alcune "epoche d'oro" del cinema__, ad esempio a metà degli anni 30 (pensiamo a "Via col vento", "Il Mago di Oz"..), a metà degli anni 70 ("Apocalypse Now", "Arancia Meccanica"..) e a metà degli anni 80 ("Scarface", "The Blues Brothers", "L'Attimo Fuggente"..). In questi periodi abbiamo avuto una grande quantità di film. Anche il numero di voti aumenta in corrispondenza di questi anni, ma in maniera meno definita. Questo potrebbe significare che non necessariamente un periodo florido per l'industria cinematografica corrisponda a un aumentato interessamento del pubblico a distanza di anni.

Al contrario, vediamo come ci sia un abbassamento notevole di numero di voti in corrispondenza della seconda guerra mondiale.


Se la curva del numero di film è abbastanza liscia, la curva del numero di voti presenta dei marginali picchi in corrispondenza di specifici anni, che balzano fuori rispetto agli anni adiacenti. Infatti, basta anche solo un film estremamente famoso che con il suo numero di voti spropositatemente grande rispetto agli altri, può far schizzare fuori un picco in corrispondenza di quell'anno.



# Conclusioni e Spunti per analisi future

### Conclusioni 

Questa analisi è una analisi preliminare e basilare, ma è in grado di dare informazioni interessanti, che possiamo riassumere principalmente in:

- il voto di un film dipende non solo dalla media dei voti, ma anche dal numero di voti ricevuti
- i film che spendono di più in budget incassano tendenzialmente di più
- i film che incassano molto solitamente sono solitamente anche ben voluti dagli utenti
- i film che spendono molto in budget non sono necessariamente più apprezzati, solo leggermente 
- gli utenti non sembrano avere preferenze per film moderni o datati
- il numero di film e di voti sono cresciuti parallelamente nel corso degli anni, in maniera esponenziale
- la quantità di film nel corso degli anni ha avuto picchi in corrispondenza di epoche d'oro del cinema (~1935, ~1975, ~1985)
